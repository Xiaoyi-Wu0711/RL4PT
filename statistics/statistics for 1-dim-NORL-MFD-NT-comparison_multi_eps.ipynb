```python
# plot_path = "../Plot/1-dim-NORL-MFD-NT/comparison/"
# if not os.path.exists(plot_path):
#     os.makedirs(plot_path)
# base_path = "../results_2/Simulation_MFD_normal/NT/A_" 
# gamma  = 1
# parameter_ls = [0, 1, 2, 3, 3.8, 4, 5, 7, "change_5_3", "change_3_5"]

plot_path = "../Plot/1-dim-NORL-MFD-NT/A_std_0_5/comparison/"
if not os.path.exists(plot_path):
    os.makedirs(plot_path)
base_path = "../results_NORL_pt_60_changed_ratio/A_" 
gamma  = 1
parameter_ls = [
                # 0,
                # 1, 
                # 2, 
                # 3, 3.8, 4, 5, 
                # 7,      
                # "change_7_3",
                # "change_3_7", 
                # "change_2_4",
                # "change_4_2", 
                # "random",
                 "std_0_5"
                 ]
```

```python
fig, ax = plt.subplots(figsize=(8,6))
plt.ylabel("Return in one episode", fontsize= 10)
plt.xlabel("Episode number", fontsize= 10)
ax.tick_params(axis='x', labelsize= 10)
ax.tick_params(axis='y', labelsize= 10)
plt.tight_layout()
# plt.ylim(-70, -40)
for model_parameter in parameter_ls: 
    rw_arr_0 = np.load(file=(base_path+str(model_parameter)+"/rw.npy"))
    gamma_arr = np.array([pow(gamma,i) for i in range(60)])
    return_arr_0 = np.sum(np.multiply(gamma_arr, rw_arr_0[:, : ]), axis=1)
    label_text = "A="+ str(model_parameter)
    plt.plot(return_arr_0, label = label_text)
# ax.legend()
# fig.savefig(plot_path+'/rw_eps.jpeg', dpi=fig.dpi)

# plt.show()
```

```python
data = []
constant_a = 1
for model_parameter in parameter_ls: 
    tt_arr_0 = np.load(file=(base_path+str(model_parameter)+"/tt.npy"))
    tt_last_5_day_arr_0 = np.load(file=(base_path+str(model_parameter)+"/tt_last_5_day.npy"))
    tt_mean_0 = np.mean(tt_arr_0, axis=1)
    rw_arr_0 = np.load(file=(base_path+str(model_parameter)+"/rw.npy"))
    sw_arr_0 = np.load(file=(base_path+str(model_parameter)+"/sw.npy"))
    pt_arr_0 = np.load(file=(base_path+str(model_parameter)+"/pt.npy"))

    a = np.ones((60))
    a[-5:] *= constant_a
    gamma_arr = np.array([pow(gamma,i) for i in range(60)])
    rw_arr_1 = rw_arr_0[:]*a
    return_arr_0 = np.sum(np.multiply(gamma_arr,rw_arr_1[:,:]), axis=1)
    label_text = "A="+str(model_parameter)
    ls = [label_text, 
          np.round(np.mean(tt_mean_0),2), np.round(np.var(tt_mean_0),2), 
          np.round(np.mean(tt_last_5_day_arr_0),2),  np.round(np.var(tt_last_5_day_arr_0),2),  
          np.round(np.mean(return_arr_0),2),  np.round(np.var(return_arr_0),2), 
          np.round(np.mean(sw_arr_0)/7500,2),  np.round(np.var(sw_arr_0)/7500,2),
          np.round(np.mean(sw_arr_0[:, -5:]/7500),2),  np.round(np.var(sw_arr_0[:, -5:]/7500),2),
          np.round(np.mean(pt_arr_0),2),  np.round(np.var(pt_arr_0),2),
          np.round(np.mean(pt_arr_0[:, -5:]),2),  np.round(np.var(pt_arr_0[:, -5:]),2)
          ]
    data.append(ls)

col_names = ["A value", "mean of AITT(1 ep)", " var of AITT(1 ep)", "mean of last-5-day AITT", "var of last-5-day AITT", "mean of return", "var of return", "mean of sw", "var of sw", "mean of last-5-day sw", "var of last-5-day sw", "mean of pt", "var of pt", "mean of last-5-day pt", "var of last 5 day pt"]

print(tabulate(data, headers=col_names))
```

```python
data = []
constant_a = 1000
for model_parameter in parameter_ls: 
    tt_arr_0 = np.load(file=(base_path+str(model_parameter)+"/tt.npy"))
    tt_last_5_day_arr_0 = np.load(file=(base_path+str(model_parameter)+"/tt_last_5_day.npy"))
    tt_mean_0 = np.mean(tt_arr_0, axis=1)
    rw_arr_0 = np.load(file=(base_path+str(model_parameter)+"/rw.npy"))
    a = np.ones((100))
    a[-5:] *= constant_a
    gamma_arr = np.array([pow(gamma,i) for i in range(100)])
    rw_arr_1 = rw_arr_0[:]*a/100
    return_arr_0 = np.sum(np.multiply(gamma_arr,rw_arr_1[:,:]), axis=1)
    label_text = "A="+str(model_parameter)
    ls = [label_text, 
          np.round(np.mean(tt_mean_0),2), np.round(np.var(tt_mean_0),2), 
          np.round(np.mean(tt_last_5_day_arr_0),2),  np.round(np.var(tt_last_5_day_arr_0),2),  
          np.round(np.mean(return_arr_0),2),  np.round(np.var(return_arr_0),2)
          ]
    data.append(ls)

col_names = ["A value", "mean of AITT(1 ep)", " var of AITT(1 ep)", "mean of last-5-day AITT", "var of last-5-day AITT", "mean of return", "var of return"]

print(tabulate(data, headers=col_names))
```
